# **Using data from karaoke sessions at the asylum.**

Here we automate the  song parsing procedure for three main languages (pt-br, fr, en), and also automate the dataframe construction.




```python
songs = pd.read_csv('/content/songs.csv')
```

```python
songs
```

```python
to_drop = songs[(songs['LYRIC'] == 'o')].index
```

```python
songs.drop(to_drop, inplace = True)
```

```python
songs
```

```python
song = []
for idx, row in songs.iterrows():
  song.append(row['LYRIC'])

parsed_songs = []
for item in song:
  lang = detect(item)
  if lang == 'pt':
    lyric = nlp_pt(item)
    parsed_songs.append(lyric)
  elif lang == 'en':
    lyric = nlp_en(item)
    parsed_songs.append(lyric)
  else:
    lyric = nlp_fr(item)
    parsed_songs.append(lyric)
```

```python
structure = {
 'song_name' : songs['SONG_NAME'],
 'song_idx' : [],
 'parsed_song': [],
 'stanzas': []
}
```

```python
for parsed_song in parsed_songs:
  structure['song_idx'].append(parsed_songs.index(parsed_song))
  structure['parsed_song'].append(parsed_song)
```

```python
for parsed_song in structure['parsed_song']:
  stanza_list = []
  for sent in parsed_song.sents:
    stanza_list.append(sent)
  structure['stanzas'].append(stanza_list)
```

```python
structure_df = pd.DataFrame(structure)
```

```python
structure_df
```

```python
def getting_feelings(song_name):

  analysis =  {
    "lyrics" : [],
    "negative" :[],
    "neutral": [],
    "positive": [],
    "compound": []
  }

  for idx, row in structure_df.iterrows():
    if song_name.lower() == row['song_name'].lower():
      for stanza in structure_df['stanzas'][idx]:
        score = s.polarity_scores(str(stanza))
        analysis['lyrics'].append(stanza)
        analysis['negative'].append(score['neg'])
        analysis['neutral'].append(score['neu'])
        analysis['positive'].append(score['pos'])
        analysis['compound'].append(score['compound'])

  analyze_df = pd.DataFrame(analysis)
  zero = analyze_df[ (analyze_df['negative'] == 0.000) & (analyze_df['neutral'] == 0.000) & (analyze_df['positive'] ==0.000)].index
  analyze_df.drop(zero, inplace=True)

  chart = sns.barplot(analyze_df)

  return [analyze_df, chart]
```

```python
getting_feelings('meu mestre')
```

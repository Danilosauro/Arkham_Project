```python
from vagalume import lyrics
```

```python
data = pd.read_csv('/content/songs_week_1.csv')
```

```python
data
```

```python
def searching_song(artist, song):
  artist_name = artist
  song_name = song

  result = lyrics.find(artist_name, song_name)

  if result.is_not_found():
      return 'Song not found'
  else:
    return result.song.lyric
```

```python
to_process = []
for idx, row in data.iterrows():
  lyric = searching_song(row['ARTIST_1'], row['SONG_1'])
  to_process.append(lyric)
```

```python
for song in to_process:
  print(song, '\n')
```
